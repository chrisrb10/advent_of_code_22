```python
# NB - making an attempt to solve all of the AOC problems in pandas
import pandas as pd
```

```python
def split_string(row):
    s = row[0]
    row[1], row[2] = s[:len(s)//2], s[len(s)//2:]
    return row

priority = {**{chr(i+96):i for i in range(1,27)}, **{chr(i+38):i for i in range(27,53)}}
```

```python
df = pd.read_table('data/input_3.txt', header=None)
```

```python
#PART 1
df = df.apply(split_string, axis=1)
df['common'] = [list(x.intersection(y))[0] for x,y in zip([set(x) for x in df[1]],[set(x) for x in df[2]])]
df['score'] = df['common'].map(priority)
print(df['score'].sum())

#PART 2
tup_list = []
for i in range(0,300,3):
    badge = list((set(df.loc[i, 0]).intersection(set(df.loc[i+1, 0]))).intersection(set(df.loc[i+2, 0])))[0]
    tup_list.append((i,badge))
    
df2 = pd.DataFrame(tup_list)
df2['score'] = df2[1].map(priority)
print(df2['score'].sum())
```
